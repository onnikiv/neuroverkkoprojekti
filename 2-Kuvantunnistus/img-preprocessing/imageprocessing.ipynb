```python
import os
from PIL import Image, ImageOps


# Enable HEIC support / APPLE not necessary
import pillow_heif
pillow_heif.register_heif_opener()

INPUT_FOLDER = "unprocessed-images"
OUTPUT_FOLDER = "processed-images"

LABEL = "knife"       # tiedostonimi

os.makedirs(OUTPUT_FOLDER, exist_ok=True)

counter = 1  # integer counter (clean filenames)

for image_file in os.listdir(INPUT_FOLDER):

    if image_file.lower().endswith(('.png', '.jpg', '.jpeg', '.heic')):

        input_path = os.path.join(INPUT_FOLDER, image_file)

        try:
            with Image.open(input_path) as img:

                # Fix orientation from phone images
                img = ImageOps.exif_transpose(img)

                # Standardize to RGB
                img = img.convert("RGB")

                # Resize for CNN input
                img = ImageOps.fit(img, (224, 224), method=Image.Resampling.LANCZOS)  # comment this line for no resizing!
                
                new_filename = f"{LABEL}-{counter}.png"
                output_path = os.path.join(OUTPUT_FOLDER, new_filename)

                # Save as PNG
                img.save(output_path, format="PNG")

                counter += 1

        except Exception as e:
            print(f"Skipping {image_file}: {e}")

print("Images standardized and renamed to png format.")
```
